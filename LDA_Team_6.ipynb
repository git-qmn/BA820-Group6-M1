```python
import ast

def extract_resp_text(resp_str):
    try:
        resp_dict = ast.literal_eval(resp_str) if isinstance(resp_str, str) else resp_str
        return resp_dict.get('text', '') if isinstance(resp_dict, dict) else ''
    except (ValueError, SyntaxError):
        return ''

df_clean_resp['clean_resp_text'] = df_clean_resp['resp'].apply(extract_resp_text)
df_clean_resp[['resp', 'clean_resp_text']].head()
```

**Analysis on 'df_clean_text' df (Users' reviews)**

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower().strip()
    tokens = text.split()
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

df_sampled_clean_text['clean_text'] = df_sampled_clean_text['text'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_df=0.95, min_df=10, stop_words='english')

text_vectorized = vectorizer.fit_transform(df_sampled_clean_text['clean_text'])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

num_topics = 5

lda = LatentDirichletAllocation(n_components=num_topics, max_iter=10, random_state=42)
lda.fit(text_vectorized)
```

```python
def print_top_words(model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        print(f"Topic #{topic_idx + 1}: {' '.join(top_words)}")

print_top_words(lda, vectorizer.get_feature_names_out())
```

Analysis on 'df_clean_resp' df (Businesses' responses)

```python
vectorizer_resp = CountVectorizer(max_df=0.95, min_df=10, stop_words='english')

text_vectorized_resp = vectorizer_resp.fit_transform(df_sampled_clean_resp['clean_resp_text'])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# intiallize this to 2 because those texts look like a thank you response to users
num_topics = 2

lda_resp = LatentDirichletAllocation(n_components=num_topics, max_iter=10, random_state=7)
lda_resp.fit(text_vectorized_resp)
```
